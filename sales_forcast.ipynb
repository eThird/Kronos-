droping store and item columns

```python
store_sales= store_sales.drop(['store','item'],axis=1)
```

converting date from object data type to datwe time data type  as it is object data type it will arise the problem of false data reading

```python
store_sales['date']= pd.to_datetime(store_sales['date'])
```

```python
store_sales.info()
```

till now we are seeing that the above model is predicting date over a sinle day so we nee to confert date onto period of month so for  to do we implement following

```python
store_sales['date']=store_sales['date'].dt.to_period("M")
monthly_sales=store_sales.groupby('date').sum().reset_index()#created newvariable for monthely sales
```

converting resulting date column to timestam data type(is used for values that contain both date and time parts)

```python
monthly_sales['date']=monthly_sales['date'].dt.to_timestamp()
```

```python
monthly_sales.head(10)
```

now visulising the data

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.xlabel("dates")
plt.ylabel("sales")
plt.title("Monthly customer sales")
plt.show()
```

call the difference on the sales column topo make sales data stationery to improve teh traing phase of the learning model

```python
monthly_sales['sales_diff']=monthly_sales['sales'].diff()
monthly_sales=monthly_sales.dropna()
monthly_sales.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.xlabel("dates")
plt.ylabel("sales")
plt.title("Monthly customer sales Difference ")
plt.show()
```

now we have to train model to tellus about next month we are going to this by getting the sales in specific month for now here we are using only 12month data and here we are droping sales and date model cuz we need only difference to train our model and reinforcment learning model later

```python
supervised_data= monthly_sales.drop(['date','sales'], axis=1)
```

preparing supervised data

```python
for i in range(1,13):
  col_name= 'month_'+ str(i)
  supervised_data[col_name]=supervised_data['sales_diff'].shift(i)
supervised_data=supervised_data.dropna().reset_index(drop=True)#method allows you reset the index back to the default 0, 1, 2 etc indexes
supervised_data.head(10)
```

spliting data into Train and test

```python
train_data= supervised_data[:-12]
test_data= supervised_data[-12:]
print("train data set :", train_data.shape)
print("test data set :", test_data.shape)
```

```python
scaler=MinMaxScaler(feature_range=(-1,1))#This estimator scales and translates each feature individually such that it is in the given range on the training set, e.g. between zero and one.
scaler.fit(train_data)
train_data =scaler.transform(train_data)
test_data=scaler.transform(test_data)
```

```python
X_train, y_train = train_data[:,1:], train_data[:,0:1]
X_test, y_test = test_data[:,1:], test_data[:,0:1]

y_train = y_train.ravel()

y_test = y_test.ravel()

print("X_train Shape: ", X_train.shape)

print("y_train Shape: ", y_train.shape)

print("X_test Shape: ", X_test.shape)

print("y_test Shape: ", y_test.shape)
```

making prediction data frame to merge the predicted sales of all traiuned algs


```python
sales_dates=monthly_sales['date'][-12:].reset_index(drop=True)
predict_df=pd.DataFrame(sales_dates)
```

```python
act_sales=monthly_sales['sales'][-13:].to_list()
print(act_sales)
```

create inear regression model and predict out put

```python
lr_model =LinearRegression()
lr_model.fit(X_train, y_train)
lr_pre = lr_model.predict(X_test)
```

```python
lr_pre = lr_pre.reshape(-1,1)
#this is a set matrix- contains the input features of the test data, and also the predicted data
lr_pre_test_set = np.concatenate([lr_pre, X_test], axis=1)
lr_pre_test_set = scaler.inverse_transform(lr_pre_test_set)
```

```python
result_list =[]
for index in range(0, len(lr_pre_test_set)):
  result_list.append(lr_pre_test_set[index] [0] + act_sales[index])
lr_pre_series = pd.Series(result_list, name="Linear Prediction")
predict_df = predict_df.merge(lr_pre_series, left_index = True, right_index= True)
```

```python
lr_mse = np.sqrt(mean_squared_error(predict_df['Linear Prediction'], monthly_sales['sales'][-12:]))
lr_mae = mean_absolute_error(predict_df['Linear Prediction'], monthly_sales['sales'][-12:])
lr_r2 = r2_score=(predict_df['Linear Prediction'], monthly_sales['sales'][-12:])
print("Linear Regression MSE: ", lr_mse)
print("Linear Regression MSE: ", lr_mae)
print("Linear Regression R2: ",lr_r2)
```

```python
plt.figure(figsize=(15,5))
# Actual sales
plt.plot(monthly_sales['date'], monthly_sales['sales'])
# Predicted sales
plt.plot(predict_df['date'], predict_df['Linear Prediction'])
plt.title("Customer sales Forecast using LR Model")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend(['Actual Sales', 'Predicted sales'])
plt.show()
```
